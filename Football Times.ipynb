```python
#Division1Femininetimes
#FAWSLtimes
#Iberdrolatimes

import pandas as pd

spain_df=pd.read_csv('D:\\FAWSLtimes.csv', encoding='latin1')
spain_df
```

```python
del spain_df['Unnamed: 7']
del spain_df['1st H.']
del spain_df['2nd H.']

#spain_df
```

```python
dataTypeSeries = spain_df.dtypes
#dataTypeSeries
```

```python
def onlyscored(str):
    dash=str.find('-')
    #scored=str[:dash]
    scored=str[dash+1:] #concede
    return int(scored)
```

```python
spain_df["0-15"] = spain_df["0-15"].apply(onlyscored)
spain_df["16-30"] = spain_df["16-30"].apply(onlyscored)
spain_df["31-45"] = spain_df["31-45"].apply(onlyscored)
spain_df["46-60"] = spain_df["46-60"].apply(onlyscored)
spain_df["61-75"] = spain_df["61-75"].apply(onlyscored)
spain_df["76-90"] = spain_df["76-90"].apply(onlyscored)
spain_df.head()
```

```python
spain_df=spain_df.set_index("Overall")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [14, 20]
for index, team in enumerate(spain_df.index):
    plt.subplot(6, 3, index+1)
    row = spain_df.iloc[index]
    row.plot(kind='bar')
    plt.title(team)
    plt.xlabel("Time Interval")
    #plt.ylabel("Goals Scored")
    plt.ylabel("Goals Conceded")
    plt.ylim(ymin=0, ymax=27)
    
plt.tight_layout()
plt.savefig('D:\FAWSLConceded.jpg')
plt.show()
```
